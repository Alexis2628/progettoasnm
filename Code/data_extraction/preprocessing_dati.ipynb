```python
df = pd.read_csv('post_Artificial_Intelligence.csv').drop_duplicates()
print(df.shape)
df
```

```python
import pandas as pd
import numpy as np

numeric_cols = ['post_pk', 'like_count', 'taken_at', 'user_pk', 'quote_count', 'repost_count', 'reshare_count']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
df['taken_at'] = pd.to_datetime(df['taken_at'], unit='s', errors='coerce')
df.dropna(subset=['id', 'post_pk', 'taken_at', 'username', 'user_pk'], inplace=True)
df[numeric_cols] = df[numeric_cols].fillna(0)
df['caption_text'] = df['caption_text'].fillna('')
df['like_count'] = df['like_count'].astype(int)
df['quote_count'] = df['quote_count'].astype(int)
df['repost_count'] = df['repost_count'].astype(int)
df['reshare_count'] = df['reshare_count'].astype(int)
df['caption_text'] = df['caption_text'].str.strip()
df.reset_index(drop=True, inplace=True)
print(df.shape)
df
```

```python
import re
df_original = df.dropna()
def remove_links(text):
    return re.sub(r'https?://\S+', '', text)
df_original['caption_text'] = df_original['caption_text'].apply(remove_links)
df_original.shape
```

```python
import pandas as pd
from deep_translator import GoogleTranslator

try:
    df = pd.read_csv('dataframe_tradotto.csv')
    print("Caricato il file con checkpoint precedente.")
except FileNotFoundError:
    df = df_original
    print("Caricato il dataframe originale.")

if 'caption_text_translated' not in df.columns:
    df['caption_text_translated'] = None

translator = GoogleTranslator(source='auto', target='en')

for index, row in df.iterrows():
    if pd.notnull(row['caption_text']) and pd.isnull(row['caption_text_translated']):
        try:
            translation = translator.translate(row['caption_text'])
            df.at[index, 'caption_text_translated'] = translation
            print(f"Riga {index} tradotta con successo.")
            if index % 10 == 0:
                df.to_csv('dataframe_tradotto.csv', index=False)
            
        except Exception as e:
            print(f"Errore alla riga {index}: {e}")
            print("Si sospende l'esecuzione. Riprova piÃ¹ tardi.")
            break

print("Traduzione completata (o interrotta in caso di errore).")
```
