```python
import pandas as pd
import numpy as np

numeric_cols = ['post_pk', 'like_count', 'taken_at', 'user_pk', 'quote_count', 'repost_count', 'reshare_count']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
df['taken_at'] = pd.to_datetime(df['taken_at'], unit='s', errors='coerce')
df.dropna(subset=['id', 'post_pk', 'taken_at', 'username', 'user_pk'], inplace=True)
df[numeric_cols] = df[numeric_cols].fillna(0)
df['caption_text'] = df['caption_text'].fillna('')
df['like_count'] = df['like_count'].astype(int)
df['quote_count'] = df['quote_count'].astype(int)
df['repost_count'] = df['repost_count'].astype(int)
df['reshare_count'] = df['reshare_count'].astype(int)
df['caption_text'] = df['caption_text'].str.strip()
df.reset_index(drop=True, inplace=True)
print(df.shape)
df
```

```python
import re
df_original = df.dropna().copy()
def remove_links(text):
    return re.sub(r'https?://\S+', '', text)
df_original['caption_text'] = df_original['caption_text'].apply(remove_links)
df_original.shape
```

```python
import pandas as pd
from deep_translator import GoogleTranslator

try:
    df = pd.read_csv('posAiT.csv')
    print("Caricato il file con checkpoint precedente.")
except FileNotFoundError:
    df = df_original
    print("Caricato il dataframe originale.")

if 'caption_text_translated' not in df.columns:
    df['caption_text_translated'] = None

translator = GoogleTranslator(source='auto', target='en')

for index, row in df.iterrows():
    if pd.notnull(row['caption_text']) and pd.isnull(row['caption_text_translated']):
        try:
            print("-------------------------------------------------------------------------------------")
            print(row['caption_text'])
            translation = translator.translate(row['caption_text'])
            print(translation)
            print("-------------------------------------------------------------------------------------")
            if translation == None:
                translation = row['caption_text']
            df.at[index, 'caption_text_translated'] = translation
            print(f"Riga {index} su {df.shape[0] }tradotta con successo.")
            if index % 1 == 0:
                df.to_csv('posAiT.csv', index=False)
            
        except Exception as e:
            print(f"Errore alla riga {index}: {e}")
            print("Si sospende l'esecuzione. Riprova più tardi.")
            break

print("Traduzione completata (o interrotta in caso di errore).")
```

```python
import os
import pandas as pd
from sentiment_analysis import SentimentAnalyzer

sa = SentimentAnalyzer(method="huggingface")

try:
    df = pd.read_csv('posAiT.csv')
    print("Caricato il file con checkpoint precedente.")
except FileNotFoundError:
    print("Caricato il dataframe originale.")
results_file = 'sentiment_results.csv'

if os.path.exists(results_file):
    sentiment_df = pd.read_csv(results_file)
    processed_indexes = set(sentiment_df['index'].tolist())
    print("Checkpoint dei risultati sentiment trovato e caricato.")
else:
    sentiment_df = pd.DataFrame(columns=['index', 'sentiment_score', 'sentiment_label'])
    processed_indexes = set()

for index, row in df.iterrows():
    if pd.notnull(row['caption_text']) and index not in processed_indexes:
        try:
            text = str(row['caption_text'])
            sentiment = sa.pipeline(text,truncation=True, max_length=512)[0]
            if translation is None:
                label = "NEUTRAL"
                score = 0
            else:
                label = sentiment["label"]
                score = sentiment["score"]
            
            new_data = {'index': index, 'sentiment_score': score, 'sentiment_label': label}
            header = not os.path.exists(results_file)
            pd.DataFrame([new_data]).to_csv(results_file, mode='a', header=header, index=False)
            
            print(f"Riga {index} su {df.shape[0]} elaborata con successo.")
            
        except Exception as e:
            print(f"Errore alla riga {index}: {e}")
            print("Si sospende l'esecuzione. Riprova più tardi.")
            break

print("Elaborazione completata (o interrotta in caso di errore).")
```
